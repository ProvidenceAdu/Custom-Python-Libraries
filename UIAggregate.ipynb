```python
import arcpy 
```

```python
class aggregate():
    
    '''
    This class executes aggregation data for Urban Institute's QOL Variables
    '''
    
    def __init__(self):
        '''
        The aggregate class is initialized with no parameters
        '''
        pass
        
    
    def withinNPA(self,NPA,AggregateFeature,OutputName):
        
        '''
        The withinNPA method aggregates all points feature classess that are completely contained by an NPA
        The with NPA method takes three parameters:
        - Input NPA which is supposed to be a polygon
        - Input feature class for aggregation which is supposed to be a point
        - Name of output feature class
        '''
        
        self.NPA = NPA
        self.AggregateFeature = AggregateFeature
        self.OutputName = OutputName
        
        print('mapping fields for ' + self.NPA +  ' and ' + self.AggregateFeature)
        
        fields = arcpy.FieldMappings()
        fields.addTable(self.NPA)
        fields.addTable(self.AggregateFeature)
        
        print('Done mapping fields')
        
        print('Executing spatial join for ' + self.NPA +'s ' + 'that completely contains '+ self.AggregateFeature)
        
        arcpy.analysis.SpatialJoin(self.NPA,
                                   self.AggregateFeature,
                                   self.OutputName,
                                  "JOIN_ONE_TO_ONE",
                                   "KEEP_ALL",fields,
                                  "COMPLETELY_CONTAINS",
                                   None, '')
        
        print('Done running spatial join')
        
    def withNPAID(self,InFeature,NPA,OutputName):
        
        '''
        The withNPAID methods assigns NPA ID to all point feature classes that are completely with an NPA
        The withNPAID method takes three parameters :
        - Input feature class which is supposed to a point 
        - Input NPA feature class which is supposed to be a polygon 
        - Name of output feature class 
        '''
        

        self.NPA = NPA
        self.InFeature = InFeature
        self.OutputName = OutputName
        
        print('mapping fields for ' + self.InFeature +  ' and ' + self.NPA)
        
        print('Executing spatial join for ' + self.InFeature + 'that are completely within '+ self.NPA)
        
        fields = arcpy.FieldMappings()
        fields.addTable(self.InFeature)
        fields.addTable(self.NPA)
        
        print('Done running spatial join')
        
        arcpy.analysis.SpatialJoin(self.InFeature,
                                   self.NPA,
                                   self.OutputName,
                                  "JOIN_ONE_TO_ONE",
                                   "KEEP_ALL",fields,
                                  "COMPLETELY_WITHIN",
                                   None, '')
```
